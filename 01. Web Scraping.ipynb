```python
# Import libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Connect to the site and pull in the data so we can have a look at the HTML. We just want the Authors and their Quotes

url = 'https://quotes.toscrape.com'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

page = requests.get(url, headers=headers)

soup1 = BeautifulSoup(page.content, 'html.parser')

soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')

title = soup2.find(style='text-decoration: none').get_text()

table = soup2.find_all('div', class_='quote')

print(title)
print(table)
```

```python
# Gather all the quotes

quotes = soup2.find_all('span', class_='text')

quotes_text = [itemprop.text.strip() for itemprop in quotes]

print(quotes_text)
```

```python
# Gather all the authors

authors = soup2.find_all('small', class_='author')

authors_text = [itemprop.text.strip() for itemprop in authors]

print(authors_text)
```

```python
# Create a Dictionary from our data and load it into a DataFrame

dict = {'Author': authors_text, 'Quote': quotes_text}

df = pd.DataFrame(dict)

print(df)
```
